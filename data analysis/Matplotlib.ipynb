## 1.绘制折线图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams    # 这一行是配置图像中的中文字体
rcParams['font.sans-serif'] = ['Microsoft YaHei']


# 创建图表，设置大小
plt.figure(figsize=(10,5))    # 这一步类似于创建画布，并不直接显示图表

# 数据
month = ['Jan.','Feb.','Mar.','Apr.']
sales = [100,150,90,130]
# 绘制折线图的方法
plt.plot(month,sales,label='产品A',linewidth=1,marker='o')
# 先填入的列表作为x轴，后填入的作为y轴，label参数会填入图例中，linewidth表示折线的粗细，marker给数据点做标记，'o'为圆点标记

# 添加标题
plt.title('2025年销售趋势',color = 'yellow',fontsize = 20)    # fontsize是字体的意思，和word里的字体大小量度好像一样

# 添加坐标轴标签
plt.xlabel('月份',fontsize = 10)
plt.ylabel('销售额（万元）',fontsize = 10)

# 添加图例
plt.legend(loc='upper left',fontsize = 10)    # loc参数表示图例的位置

# 添加网格线
plt.grid(axis='y',alpha=0.7,linestyle='--')    # axis:显示哪条轴上的网格线；alpha:线条透明度；linestyle:实线为'-',虚线为'--'
                                               # 默认x,y轴都有            默认为1          默认为实线

# 设置刻度字体大小
plt.xticks(rotation=0,fontsize = 10)
plt.yticks(rotation=-10,fontsize = 8)    # rotation参数可以表示字体的旋转角度。其中正为逆时针旋转，负为顺时针旋转

# 设置y轴的范围
plt.ylim(0,200)

# 在每个数据点上显示数值
for x,y in zip(month,sales):
    plt.text(x,y+5,str(y),ha='center',va='bottom',fontsize = 10)
# 其中前两个参数是数值显示的坐标，因此y+5就是在数据点的坐标往上移了五个单位；str(y)指显示的内容为对应的纵坐标；
# ha指水平方向上折线相对于数值的偏移程度，'center'不偏移，'left'左偏,'right'右偏；
# va则是垂直方向上折线的相对位置，'bottom'则折线在数值的下方；'top'相反

# 显示图表
plt.show()


# 怎么感觉比Excel更复杂？（疑惑
```

## 2.条形图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams    # 这一行是配置图像中的中文字体
rcParams['font.sans-serif'] = ['Microsoft YaHei']


# 创建图表，设置大小
plt.figure(figsize=(10,5))    # 这一步类似于创建画布，并不直接显示图表

# 数据
subjects = ['Math','English','Physics','Chemistry']
scores = [85,92,78,88]

# 绘制柱状图
plt.bar(subjects,scores,
        label='Score',
        width=0.5)    # width表示柱状图的宽度

# 标题和图例
plt.title('bro的期末成绩',fontsize = 20,color= 'red')
plt.legend(loc='upper left',fontsize = 10)

# 添加网格线
plt.grid(axis='y',alpha=0.7,linestyle='--')
plt.ylim(0,100)

# 自动优化排版
plt.tight_layout()

# 添加标注
for x,y in zip(subjects,scores):
    plt.text(x,y+1,str(y),ha='center',fontsize = 10)
plt.show()
```

## 3.饼图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams    # 这一行是配置图像中的中文字体
rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.figure(figsize=(10,5))

# 要绘制的数据
payment_kind = ['food','book','traffic','others']
money = [600,100,50,250]
# colors=['#66b3ff#,...]    可以这样设置各个扇形的颜色。想要自己喜欢的颜色就去找个调色板看名字。这个list作为参数color的参数值填入绘制方法中
# 绘制饼图
plt.pie(money,labels=payment_kind,autopct='%1.1f%%',startangle=90,pctdistance=0.6)
# 注意这里标注被绘制对象的参数是labels，不要忘了s；autopct可以给饼图的各个扇形中加上百分比，
# startangle可以体哦阿正初始化图的角度，pctdistance调整百分比数字的位置（沿半径方向）

# 标题
plt.title('每月开销',fontsize = 20,color= 'orange')

plt.tight_layout()

plt.show()
```

```python
# 环形图的绘制
# 只需要在饼图绘制的基础上加上参数

import matplotlib.pyplot as plt
from matplotlib import rcParams    # 这一行是配置图像中的中文字体
rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.figure(figsize=(10,5))

# 要绘制的数据
payment_kind = ['food','book','traffic','others']
money = [600,100,50,250]
# colors=['#66b3ff#,...]    可以这样设置各个扇形的颜色。想要自己喜欢的颜色就去找个调色板看名字。这个list作为参数color的参数值填入绘制方法中
# 绘制饼图
plt.pie(money,labels=payment_kind,autopct='%1.1f%%',startangle=90,wedgeprops={'width':0.6})
# wedgeprops表示中间空出来的圆的半径

# 在圆环中心添加标题
plt.text(0,0,'总计：\n100%',ha='center',fontsize=10)    # 前两个是标题位置的极坐标

# 标题
plt.title('每月开销',fontsize = 20,color= 'orange')

plt.tight_layout()

plt.show()
```

```python
# 如果想要给环形图添加上突出块，可以：
import matplotlib.pyplot as plt
from matplotlib import rcParams    # 这一行是配置图像中的中文字体
rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.figure(figsize=(10,5))

# 要绘制的数据
payment_kind = ['food','book','traffic','others']
money = [600,100,50,250]
# colors=['#66b3ff#,...]    可以这样设置各个扇形的颜色。想要自己喜欢的颜色就去找个调色板看名字。这个list作为参数color的参数值填入绘制方法中
explode = [0,0.2,0,0]    # 设置突出块的位置和突出程度

# 绘制饼图
plt.pie(money,labels=payment_kind,autopct='%1.1f%%',startangle=90,pctdistance=0.6,explode=explode,shadow=True)
# 注意这里标注被绘制对象的参数是labels，不要忘了s；autopct可以给饼图的各个扇形中加上百分比，
# startangle可以体哦阿正初始化图的角度，pctdistance调整百分比数字的位置（沿半径方向）

# 标题
plt.title('每月开销',fontsize = 20,color= 'orange')

plt.tight_layout()

plt.show()
```

## 4.散点图

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams    # 这一行是配置图像中的中文字体
rcParams['font.sans-serif'] = ['Microsoft YaHei']


# 创建图表，设置大小
plt.figure(figsize=(10,10))    # 这一步类似于创建画布，并不直接显示图表

# 数据，模拟线性回归
np.random.seed(0)
x = list(np.random.randint(0,100,500))
y = list([np.random.uniform(max(0, xi-10), min(100, xi+10)) for xi in x])    # 这玩意好像叫列表推导式，比原始的for循环更简洁

# 绘制散点图的方法
plt.scatter(x,y,marker='o')

plt.plot([0,100],[0,100],color = 'red')
# 根据所选区间内的点生成回归曲线:plt.plot([x1,x2],[y1,y2])，这样写就明白了

plt.show()


# 怎么感觉比Excel更复杂？（疑惑
```

## 5.箱线图

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams    # 这一行是配置图像中的中文字体
rcParams['font.sans-serif'] = ['Microsoft YaHei']

# 箱线图是什么，看过论文的都知道（畏惧
np.random.seed(0)
data={
    '语文':np.random.randint(70,100,10),
    '数学':np.random.randint(65,90,10),
    '英语':[88,83,86,83,85,94,90,89,28,93]    # 创造一个异常值
}
plt.figure(figsize=(10,5))

plt.boxplot(data.values(), tick_labels=data.keys())

plt.title('各科成绩分布图')
plt.ylabel('分数')
plt.ylim(0,100)
plt.grid(axis='y',alpha=0.7,linestyle='--')

plt.show()
```

### 如果要同时画多个图？

```python
month = ['Jan','Feb','Mar','Apr','May','Jun']
sales = [100,120,130,120,150,90]
f1 = plt.subplot(2,2,1)    # 这里是确定这个图在哪个位置，前面的两个2是指整个画布有2×2=4个位置，后面的1是指它占据第一个位置
f1.plot(month,sales)
f2 = plt.subplot(2,2,3)
f2.bar(month,sales)
f3 = plt.subplot(2,2,4)
f3.scatter(month,sales)
```
