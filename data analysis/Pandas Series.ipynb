# Pandas Series

```python
import pandas as pd
import numpy as np
```

## 1.Series的创建

```python
series_1_1 = pd.Series([1,2,3,4,5,6]) # pd.Series中一定要大写
print(series_1_1)
# 可以发现，打印结果中，第一列是索引，第二列才Series的值

# 自定义索引
series_1_2 = pd.Series([100,200,300,400,500,600], index=["A","B","C","D","E","F"])
print(series_1_2)  # 这样看来，pandas和实际应用更接近了

# 自定义name
series_1_4 = pd.Series([100,200,300,400,500,600], index=["A","B","C","D","E","F"], name="Price")
print(series_1_4)  # name可以用于描述Series中数据的含义
```

```python
# 通过字典来创建
series_1_3 = pd.Series({"Jay":18, "Sue":17, "Jimi":27, "V":20, "Tom":45})
print(series_1_3) # 和index异曲同工，相比之下index方便一点点

# index还可以拿来当索引使用（毕竟它本身就是索引对象
series_1_4 = pd.Series(series_1_3,index=["Jay","Sue"])
print(series_1_4)
```

## 2.Series的属性

```python
print(series_1_3)
print(series_1_3.index)
print(series_1_3.values) # Series的值
print(series_1_3.shape)
print(series_1_3.size) # 元素个数
print(series_1_3.ndim)
print(series_1_3.dtype)
series_1_3.name = "Age"
print(series_1_3.name) # Series的名称
print(series_1_3.loc["Jay"]) # 显示索引，用index作为索引，输出结果为value。这是和前面用index的区别。
print(series_1_3.loc["Jay":"V"]) # 显示索引还可以支持切片形式,这样的输出结果是Series格式
print(series_1_3.iloc[0:2]) # 隐式索引，用的是从0开始的整数，也就是最原始的索引。
                            # 看到这里要注意，显示索引包含右边。（搞这么乱七八糟要干嘛，不能统一吗
print(series_1_3.at["Jay"])
print(series_1_3.iat[4]) # 这两个都不能用切片，只能调用一个元素。


```

```python
# 访问series数据的方式
series_1_5 = pd.Series([1,2,3,4,5,6],index=['A','B','C','D','E','F'])
print(series_1_5[1]) # 这种索引方式虽然不会报错，但会被pycharm警告。
print(series_1_5["A"]) # 可以直接用index部分作为索引调用数据
print(series_1_5[series_1_5 < 4]) # 还可以使用布尔索引，输出值为series格式
print(series_1_5.head()) # .head(num)可以查看前num行数据，不输入默认5行。.tail()类似
print(series_1_5.tail(3))
```

## 3.Series常见函数

```python
series_3_1 = pd.Series([1,2,np.nan,4,4,4,None,6],index=['A','B','C','D','E','F','G','H'])
print(series_3_1)
```

```python
series_3_1.head() # 可以直接执行.head()方法，而且输出结果还可以呈现表格或者视图
```

```python
# 查看所有的描述性信息，非缺失值的元素个数，平均值，标准差，极值，四分位数
# series_3_1.describe()       这条方法也可以直接跑，但是如果cell里有了其他代码，它就不跑了。这是为什么呢
#                                                答：一个 cell 只会自动显示“最后一行表达式”的返回值

print(series_3_1.count())  # 想要获取非缺失值的元素个数
print(series_3_1.keys())
print(series_3_1.index)  # 二者都可以获取index，只是.keys()是方法，.index是属性
print(series_3_1.isna())  # 检查否为缺失值，输出布尔值Series
print(series_3_1.isin([2,3,5,7,11,13]))  # 检查series元素是否在某一个数组（还是Series?)中，也是输出布尔值Series
```

```python
# 下面这些的含义都是和NumPy一样的，但调用方法有所区别
print(series_3_1.mean())
print(series_3_1.median()) # 中位数
print(series_3_1.min())
print(series_3_1.max())
print(series_3_1.std())
print(series_3_1.var())
print(series_3_1.sum())
# 有区别的是分位数
print(series_3_1.quantile(0.25))  # 在NumPy里输入的是25

# 在NumPy里没出现过的：众数
print(series_3_1.mode())
```

```python
series_3_1.sort_values()  # 按照值排序
# 默认为从低到高排序。如果要从高到低，则：
series_3_1.sort_values(ascending=False)   # 相当于括号里什么都不写默认ascending为True
print(series_3_1.sort_index())   # 按照索引排序
```

```python
# 去重
print(series_3_1.drop_duplicates())  # 返回的Series仍然会保留原Series中的index和元素对应关系
```

```python
# 取差值，取每个值和前一个值的插值，第一个为NaN
series_3_2 = pd.Series([28,38,34,23,34,23,27])
print(series_3_2.diff())
```

#### （至此应该发现，在Series里，index很重要。）

```python
# 特殊的索引方法
price = pd.Series([102.3,103.5,105.1,104.8,106.2,107.0,106.5,108.1,109.3,110.2],
                  index = pd.date_range('2023-01-01',periods=10))
# Series为以日期为索引专门设计了一个方法。periods可以理解为一共有多少天
print(price)
# Pandas还专门有一个计算Series里每日收益率的方法
print(price.pct_change())
# 找出哪一天收益最高
print(price.pct_change().argmax())  # 这是NumPy里的方法，结果是直接的普通索引（也就是从0开始的数字），不如index直观
print(price.pct_change().idxmax())  # 这样输出结果就是最大值对应的index了
print(price.pct_change().idxmin())
```

```python
sales = pd.Series([120,135,145,160,155,170,180,175,190,200,210,220],
                  index = pd.date_range('2022-01-01',periods = 12,freq = 'MS') )
# freq参数可以选则日期时的单位，'D'为天，'W'为周，'MS'为月
# 另外月还有'ME'这一种单位，貌似'MS'月初，'ME'代表月末，这会影响到后续输出索引（对a的赋值），具体是为什么呢？
# 每个季度的平均销量
print(sales.resample('QS').mean()) # resample()可以更换日期单位对Series进行重新采样
a = sales.pct_change()
print(a[a>0].index)
b = a>0
print(b)
print(b.rolling(3).sum()) # 这是滚动窗口操作，相当于对每连续三个月的布尔值求和（0和1）（具体操作由你的计算函数决定）
```
