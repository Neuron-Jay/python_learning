# Pandas DataFrame

```python
import pandas as pd
import numpy as np
```

## 1.DataFrame的创建

```python
# 通过Series来创建
column_1 = pd.Series([1,2,3,4,5])
column_2 = pd.Series([6,7,8,9,10])
df_1_1 = pd.DataFrame({"第1列":column_1,"第2列":column_2}) # 这里是以字典的格式去创建一个DataFrame
print(df_1_1)
# 通过字典创建
df_1_2 = pd.DataFrame(
    {
        "id":[1,2,3,4,5],
        "name":['a','b','c','d','e'],    # 这里的字典好像已经不是简单一对一的键值对了
        "age":[20,22,23,24,25]
        # 你在这里还可以用同样方法加更多的信息类型
    },index=[1,2,3,4,5]
)
print(df_1_2)

# DataFrame可简单当作几个Series拼接起来，更像一个Excel表格了，或者说二维数组的应用化
```

## 2.DataFrame的属性

```python
# DataFrame的属性基本和Series一致

# columns, 列标签
print(df_1_2.columns)
# 值
print(df_1_2.values)    # 输出结果是一个二维数组，可想而知它的ndim也是2
print(df_1_2.dtypes)    # 在DataFrame里还是复数，不能用dtype，得加s
# 行列转置
print(df_1_2.T)
```

```python
# 获取元素

# 获取某行
print(df_1_2.loc[3]) # 用index来引用
print(df_1_2.iloc[3]) # 用从0开始的整数，也就是最原始索引来引用
# 获取某列
print(df_1_2.loc[:,'age'])  # 由于是根据第二个维度(axis = 1)上的columns获取某列，所以得用二维索引的格式，
                             # 为什么这里[:,'age']中前面的冒号不能像数组里一样可省略呢？
                             # 因为loc[]是标签索引，必须有明确的行标签与列标签的位置
                             # 上面那个如果省略冒号:[,'age'],
# 或者直接用columns
print(df_1_2['age'])

#获取单个元素
print(df_1_2.at[3,'age'])
print(df_1_2.iloc[2,2])
```

```python
# 随机抽样
print(df_1_2.sample())
```

## 3.DataFrame的方法

```python
# 很多和Series一样

# 示例DataFrame
df_3 = pd.DataFrame(
    {
        "id":[1,2,3,4,5],
        "name":['jack','ross','mary','plant','john'],
        "age":[20,22,23,24,25],
        "scores":[78,92,89,67,83]
    },index=[1,2,3,4,5]
)
print(df_3)
# .head(), .tail()和Series一样

```

```python
print(df_3.isin(['jack',20,3])) # 查看要找的元素（可有多个不同类型的元素）是否在该DataFrame中，会对所有的元素进行判断
```

```python
print(df_3.sum())    # 它会把所有列都分别进行求和，但有些列的求和没有意义
print(df_3['scores'].mean())    # 对选定的某一列进行统计学运算
print(df_3.scores.var())    # 还可以直接这样写
# 如果要选定多列，则要：
df_3[['age','scores']]    # 这样就可以同时对多个数据列进行统计学运算
                          # 为什么要打两个中括号呢？第一层是索引，第二层是表示列表
```

```python
df_3_1 = pd.DataFrame(
    {
        "id":['a','a','b','c','d','e'],
        "name":['jack','jack','ross','mary','plant','john'],
        "age":[20,20,22,23,24,25],
        "scores":[78,78,92,89,67,83]
    },index=[1,2,3,4,5,6]
)
print(df_3_1)
print(df_3_1.duplicated())    # 可以检测是否有重复值。只有所有columns里的元素都相同才会被识别为重复，输出True
print(df_3_1.duplicated(subset=['age']))     # 对某一列进行重复检测
```

```python
# 排序
# 按索引排序，和Series方法相同
# 按值排序
print(df_3.sort_values(by = 'scores', ascending = False))    # 要按照某一column(分数)排序，得添加参数by=(column)
# 如果碰到分数相同的人呢？那么可以再按照别的column来排序，比如再按照age
print(df_3.sort_values(by = ['scores','age'], ascending = [False,True]))
```

```python
# 取按照某一column排序前对象
# print(df.sort_values(by = [(column)], ascending=False).head(num)
# 实际上还可以用这个函数：
# print(df.nlargest(num, columns=[(column)])
print(df_3.nlargest(2, columns='scores'))
```

## 4.DataFrame的相加

```python
df_a = pd.DataFrame(
    {"score": [90, 85, 88]},
    index=["Tom", "Jack", "Mary"]
)

df_b = pd.DataFrame(
    {"score": [80, 70, 95]},
    index=["Jack", "Mary", "Lucy"]
)

print(df_a)
print(df_b)

print(df_a + df_b)

df_c = pd.DataFrame(
    {"math": [90, 85, 88]},
    index=["Tom", "Jack", "Mary"]
)

df_d = pd.DataFrame(
    {"english": [80, 70, 95]},
    index=["Tom", "Jack", "Mary"]
)

print(df_c + df_d)

```

由此可以看出，如果两个DataFrame相加，只有行标签index和列标签columns都对齐（一致）了，才会进行相加。

如果没对齐，会直接输出缺失值，这在实际应用中其实是一种保护措施。

```python
df = pd.DataFrame(
    {
        "math":      [88, 59, 76, 91, 67, 84, 73, 59],
        "english":   [92, 61, 70, 85, 58, 79, 88, 60],
        "physics":   [75, 55, 68, 94, 72, 81, 77, 58],
        "chemistry": [85, 62, 74, 90, 65, 87, 80, 59]
    },
    index=["Tom", "Jack", "Mary", "Alice", "Bob", "David", "Lisa", "John"]
)
df.loc[df["math"] < 60] += 10
print(df)
```
