## 2.缺失值的处理

```python
# 查看是否为缺失值的方法前面已经学过
df_2 = pd.DataFrame([[1,2,np.nan],[4,5,6],[7,None,pd.NA]],index=[1,2,3],columns=['A','B','C'])
print(df_2)
print(df_2.isna().sum())    # 这样可以查看缺失值的个数。在df里默认按照axis=0查看每列的。可以改成.sum(axis=1)
```

```python
# 剔除缺失值
print(df_2.dropna())    # 这样会剔除包含缺失值的行
                        # 很奇怪的一点，不输入axis参数还是默认axis=0，但是结果却是按照行来保留的，这是为什么？
# axis=0的具体含义是沿着竖轴进行操作，在df里就是沿着index，因此是一行一行地进行操作。同理，axis=1对应column

# 可以通过添加how参数来判断剔除的条件
print(df_2.dropna(how='all'))    # 只有这一行都是缺失值才把这一行剔除，默认how='any'
print(df_2.dropna(thresh=2))    # 如果至少有num个非缺失值，就保留
print(df_2.dropna(subset=['B','C']))    # 对某列进行缺失值剔除
```

## 4.数据变形

```python
data_4={
    'ID':[101,102],
    'name':['Alice','Bob'],
    'math':[90,85],
    'English':[88,92],
    'Science':[95,89]
}
df_4=pd.DataFrame(data_4)
print(df_4)
```

```python
# 最简单的变形：转置
print(df_4.T)
```

```python
# 使用pd.melt()函数可以对df进行如下变形，把宽表变成长表，从而把原本的列名变成数据。
df_4_1=pd.melt(df_4,id_vars=['ID','name'],var_name='科目',value_name='分数')
# 宽表在分组统计中常用，且容易被机器识别。其中各个参数：
# id_vars → 保留的列，不做变化（ID和name）
# var_name → 原列名变成哪一列（各个学科变成"科目"这一列，原来的'math''English'等作为数据填入这一列）
# value_name → 原数值变成哪一列，各科成绩填入"分数"这一列

df_4_2=df_4_1.sort_values('name')
print(df_4_2)
```

```python
# pd.pivot, 则是melt的逆操作，把长表变成宽表。
df_4_3=pd.pivot(df_4_1,index=['ID','name'],columns='科目',values='分数')
# index和columns很好理解，values是指用原先的宽表里的哪一列中的数据来填充
print(df_4_3)
```

```python
# 分列操作
data_4={
    'ID':[101,102],
    'name':['Alice Smith','Bob Smith'],
    'math':[90,85],
    'English':[88,92],
    'Science':[95,89]
}
df_4=pd.DataFrame(data_4)
df_4[['First_name','Last_name']]=df_4['name'].str.split(" ",expand=True)
                                             # ^ 这里的.str是指把对象当作字符串进行处理
# 其中expand参数的含义
# expand=False（默认）：返回 Series，每个元素是一个列表 ['Alice', 'Smith']
# expand=True：返回 DataFrame，将列表元素拆成多列
print(df_4)
```

## 5.数据分箱

```python
df_5=pd.read_csv('data/student_academic_performance.csv')
df_5.head(25)
```

```python
df_5_1 = df_5.head(10)[['Student ID','Math Score']]
df_5_1
```

```python
# 使用pd.cut()来进行分箱操作
pd.cut(df_5_1['Math Score'],bins=2)
# bins如果是一个整数num，就代表着把指定的某一列所有数值数据分为num个区间，输出结果为各个原数据所在的区间。
# 其中起始值和结束值为所有数据的min（实际上会比min小一点点）和max
pd.cut(df_5_1['Math Score'],bins=[40,50,60,70,80,90],labels=['F','E','D','C','B'])
# bin也可以这样写成list，同样表示分箱的各个区间
# labels就是给每个区间起的名字。如果有labels参数，输出结果就会是labels，可读性更高。
# 需要注意的是，labels的数据类型是 Categorical 类型。

# 分完箱以后，我们可以把分箱结果作为新的列填回原表
df_5_1['Math Score Level']=pd.cut(df_5_1['Math Score'],bins=[40,50,60,70,80,90,100],labels=['F','E','D','C','B','A'])
df_5_1
```

有关Categorical这一数据类型的介绍

对于一些在数据分析中数量有限且重复出现的数据，例如成绩分级A-F，性别，病期I-IV期等等，如果全部使用字符串即object来表示，会占用过多性能

通过astype()方法把这些数据变为pandas.Categorical类型，可以有效降低性能损耗，原理类似于把字符串比较转变为更简单的整数比较

需要注意的是，对于Category这一类型的处理，pandas有专门的一套方法名，因此在套用其他一些普适性比较广的方法名（如去重去NaN等）时请先考虑是否会导致原有的Category造成变化。

具体的方法请见《利用Python进行数据分析》的233页表7-7

```python
# pd.qcut()可以进行等频率的分割，在本案例中会让每个区间里的人数相等
pd.qcut(df_5_1['Math Score'],2)    # 这个整数参数同样表示分为几个区间
pd.qcut(df_5_1['Math Score'],2).value_counts()    #.value_counts()可以算出每个区间有多少个数据，和后面groupby的.size()很像
```

```python
df_5_2=df_5.head(25)[['Student ID','Math Score']]

df_5_2['Math_Score_Level']=pd.cut(df_5_2['Math Score'],bins=[0,50,60,70,80,90,100],labels=['F','E','D','C','B','A'])
# 可以发现，用list来表示bins参数的区间不一定需要每个区间等长
df_5_2
```

## 7.分组聚合

```python
# df.groupby('分组的字段')['聚合的字段'].聚合函数()
df_7 = pd.read_csv('data/student_academic_performance.csv')

# 先看看这个csv的基本情况
#print(df_7.isna().sum())
#print(df_7.dtypes)

# 剔除空值，并且对一些数据进行适当的转化以节约性能
df_7 = df_7.fillna({
    'Parental Education':'UNKNOWN',
    'Lunch Type':'UNKNOWN',
    'Study Time per Week':df_7['Study Time per Week'].mean(),
    'Daily Sleep Duration':df_7['Daily Sleep Duration'].mean(),
    'Reading Score':df_7['Reading Score'].mean(),
    'Writing Score':df_7['Writing Score'].mean()
})    # 剩下两个列的数据种类为布尔值，这两个列里的缺失值该如何处理呢？

df_7[['Math Score','Reading Score','Writing Score']] = df_7[['Math Score','Reading Score','Writing Score']].astype('int64')

df_7[['Gender','Race/Ethnicity','Parental Education','Lunch Type','School Type']] = df_7[['Gender','Race/Ethnicity','Parental Education','Lunch Type','School Type']].astype('category')

#print(df_7.groupby(['Parental Education']).size())    # .size()方法可以返回一个含有各个分组大小的Series，书上说很实用（思考
# 另外这一行还有FutureWarning

# 这个分组聚合具体怎么使用呢？其实在我们这个案例里就非常合适。
# 比如，根据父母的受教育程度以及学校类型不同，我们可以对受测试的学生进行分组，并按照分组计算不同分组的成绩平均值
# 可以通过列表一次性传入多个分组的参考列名，专业叫法为：键
grouped1 = df_7.groupby(['Parental Education','School Type'],observed=True)['Math Score']
# 参数observed是什么意思？当你对categorical列分组时：Pandas默认行为（observed=False)是：即使某个类别在数据中没有出现，也会显示出来。
#print(grouped1.mean())
#print(df_7.dtypes)

# 如果要统计比较多的值，例如按照前面的分组，给Math Score算平均值，给Internet Access算众数。可以使用.agg({'列名':'方法名',...})
grouped2 = df_7.groupby(['Parental Education','School Type'],observed=True).agg({
    'Math Score':'mean',
    'Reading Score':'mean',
    'Internet Access':'mean'    # 还记得True=1,False=0吗？对这个布尔Series求平均值，就可以算出True在当中所占的比例了
})
print(grouped2)
```
